```python
from gmdb.event_summary import EventSummary
```

# EventSummary


Start by creating an event summary instance by pointing to a directory of files.

```python
data_dir = "../tests/data"
event = EventSummary.fromFiles(data_dir)

# Available stations
print('AVAILABLE STATIONS:',event.stations)
```

### Once an event object has been created you can view the PGM values for each station by creating a dataframe

```python
event.getStationDataframe('EAS')
```

### You can also view a summary of all stations

```python
event.getFlatfileDataframe()
```

### Default imcs and imts can be overriden

```python
data_dir = "../tests/data"
event2 = EventSummary.fromFiles(data_dir, ['channels'], ['PGA', 'PGV', 'SA(2.0)', 'SA(0.1)'])
```

```python
event2.getStationDataframe('EAS')
```
